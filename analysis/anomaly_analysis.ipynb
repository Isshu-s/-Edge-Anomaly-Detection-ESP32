```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, axes = plt.subplots(3, 1, figsize=(13, 9), sharex=True)
fig.suptitle('Edge Anomaly Detection — ESP32 + DHT22\n(Z-score method, threshold = 2.5σ)',
             fontsize=14, fontweight='bold')

# ── Plot 1: Temperature + mean ───────────────────────
ax = axes[0]
ax.plot(df['TIME_S'], df['TEMP_C'],  color='steelblue', lw=1.3,
        label='Measured temp (°C)')
ax.plot(df['TIME_S'], df['MEAN_C'],  color='orange',    lw=1,
        ls='--', label='Rolling mean')
for _, row in df[df['FAULT']==1].iterrows():
    ax.axvspan(row['TIME_S']-0.25, row['TIME_S']+0.25,
               alpha=0.2, color='red')
ax.set_ylabel('Temperature (°C)')
ax.legend(loc='upper right', fontsize=10)
ax.set_title('Sensor readings with fault windows (shaded red)')
ax.grid(True, alpha=0.3)

# ── Plot 2: Z-score ──────────────────────────────────
ax = axes[1]
ax.plot(df['TIME_S'], df['ZSCORE'], color='purple', lw=1.3,
        label='Z-score')
ax.axhline(2.5, color='red', ls='--', lw=1.5, label='Threshold Z=2.5')
ax.fill_between(df['TIME_S'], df['ZSCORE'], 2.5,
                where=(df['ZSCORE'] > 2.5),
                alpha=0.25, color='red', label='Anomaly region')
ax.set_ylabel('Z-score')
ax.legend(loc='upper right', fontsize=10)
ax.set_title('Z-score (spikes above red line = fault detected)')
ax.grid(True, alpha=0.3)

# ── Plot 3: Fault flag ───────────────────────────────
ax = axes[2]
ax.fill_between(df['TIME_S'], df['FAULT'], alpha=0.8, color='crimson',
                step='post', label='Fault flag')
ax.set_ylabel('Fault (0/1)')
ax.set_xlabel('Time (seconds)')
ax.set_ylim(-0.1, 1.3)
ax.legend(loc='upper right', fontsize=10)
ax.set_title('Binary fault output (what the MCU outputs to LED/OLED)')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('anomaly_timeseries.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved → anomaly_timeseries.png (download from Files panel on left)")
```

```python
import numpy as np
import matplotlib.pyplot as plt

signal = df['TEMP_C'].values
N  = len(signal)
fs = 2.0   # 2 samples per second (500ms loop)

# Compute FFT
fft_vals = np.fft.rfft(signal - np.mean(signal))
fft_mag  = np.abs(fft_vals) / N
freqs    = np.fft.rfftfreq(N, d=1.0/fs)

# Find peak (skip DC at index 0)
peak_idx = np.argmax(fft_mag[1:]) + 1
peak_f   = freqs[peak_idx]
peak_m   = fft_mag[peak_idx]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle('FFT Analysis — validates Z-score threshold selection',
             fontsize=13, fontweight='bold')

# Left: full spectrum
ax1.plot(freqs, fft_mag, color='teal', lw=1.3)
ax1.axvline(peak_f, color='red', ls='--', lw=1.5,
            label=f'Peak @ {peak_f:.3f} Hz')
ax1.set_xlabel('Frequency (Hz)')
ax1.set_ylabel('Magnitude')
ax1.set_title('Frequency spectrum of sensor signal')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Right: fault vs normal temperature distributions
normal_temps = df[df['FAULT']==0]['TEMP_C']
fault_temps  = df[df['FAULT']==1]['TEMP_C']
ax2.hist(normal_temps, bins=12, color='steelblue', alpha=0.7,
         label='Normal readings')
ax2.hist(fault_temps,  bins=8,  color='crimson',   alpha=0.7,
         label='Fault readings')
ax2.axvline(normal_temps.mean() + 2.5*normal_temps.std(),
            color='black', ls='--', lw=1.5, label='2.5σ boundary')
ax2.set_xlabel('Temperature (°C)')
ax2.set_ylabel('Count')
ax2.set_title('Temp distribution — normal vs fault')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('fft_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print(f"Dominant fault frequency: {peak_f:.4f} Hz (magnitude: {peak_m:.4f})")
print("Saved → fft_analysis.png")
```
